```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
# Access maps with unique API key
gmaps.configure(api_key=g_key)
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
weather_data_path = "Weather Data.csv"
weather_data = pd.read_csv(weather_data_path)
weather_data.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Store latitude and longitude in locations
locations = weather_data[["Lat", "Lng"]]
humidity = weather_data["Humidity"].astype(float)
```

```python
# Plot Heatmap
fig = gmaps.figure(center=(45.0,0.0), zoom_level=2)

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations,weights=humidity,dissipating=False,max_intensity=np.max(humidity),point_radius=2)

# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Filter city names based of the following critiea
vacation_city = weather_data.loc[(weather_data["Wind Speed"] <= 10)
                                 &(weather_data["Wind Speed"] >= 5)
                                 &(weather_data["Cloudiness"] >= 50)
                                 &(weather_data["Maximum Temperture"] >= 25)
                                 &(weather_data["Maximum Temperture"] <= 35)
                                 &(weather_data["Humidity"] <= 75)].dropna()
# Display Vacation City Names
vacation_city
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# Create a hotel
hotel_df = vacation_city.loc[:,["City","Country", "Lat", "Lng"]]

# Add a "Hotel Name" column to the DataFrame.
hotel_df["Hotel Name"] = ""

# Display the result
hotel_df
```

```python
#Url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# Identify Parameters
params = {"type" : "hotel",
          "keyword" : "hotel",
          "radius" : 5000,
          "key" : g_key}
```

```python
print("Searching.....")   

print("-------------------------------------------------------------------------------")   

for city, row in hotel_df.iterrows():
    # get city name, lat, lnt from df
    lat = row["Lat"]
    lng = row["Lng"]
    city_name = row["City"]
    
    params["location"] = f"{lat},{lng}"
    
    # Create URL and make a request for Hotels
    print(f"Retrieving Results for Index {city}: {city_name}.")
    response = requests.get(base_url, params=params).json()
    response
    
    # Extract Hotels (Results)
    hotel = response['results']
    
    # save the hotel name to dataframe
    try:
        print(f"Closest hotel in {city_name} is {hotel[0]['name']}.")
        hotel_df.loc[city, "Hotel Name"] = hotel[0]['name']

    # if there is no hotel available, show missing field
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("-------------------------------------------------------------------------------")   
 

print("End of Search")
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content = hotel_info)

# Add the layer to the map
fig.add_layer(markers)

# Display figure
fig
```
